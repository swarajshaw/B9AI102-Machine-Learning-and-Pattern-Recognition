```python
company_name =[]
for company in soup.findAll('div',{'class':'d-flex justify-content-between align-items-start'}):
    company_name.append(company.a.text.strip())

company_name
```

```python
job_name =[]
for name in soup.findAll('div',{'class':'d-flex flex-column pl-sm css-1buaf54 job-search-key-1mn3dn8 e1rrn5ka0'}):
    job_name.append(name.findAll('a')[1].text.strip())

job_name
```

```python
job_location =[]
for loc in soup.findAll('div',{'class':'d-flex flex-wrap job-search-key-1m2z0go e1rrn5ka2'}):
    job_location.append(loc.span.text.strip())

job_location
```

```python
salary_estimate =[]
for salary in soup.findAll('div',{'class':'d-flex flex-wrap job-search-key-1a46cm1 e1rrn5ka3'}):
    salary_estimate.append(salary.span.text.strip())

salary_estimate
```

```python
job_age =[]
for age in soup.findAll('div',{'class':'mx-xxsm css-1s8337r job-search-key-crgfuk eigr9kq2'}):
    job_age.append(age.div.text.strip())

job_age
```

```python
links = []
for i in soup.findAll('div',{'class':'d-flex justify-content-between align-items-start'}):
  link = 'https://www.glassdoor.co.in'+ i.a['href']
  links.append(link) 

links
```

```python
description = []

for link in links:
  page = requests.get(link,headers=headers)
  soup1 = BeautifulSoup(page.content,'lxml')
  for job in soup1.findAll('div',{'id':'JobDescriptionContainer'})[0]:
    description.append(job.text.strip())

description
```
